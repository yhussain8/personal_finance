# Analysis of CIBC Visa credit card expenses between Jan 1, 2022 - May 3, 2023
As downloaded on May 3, 2023

```python
import pandas as pd
```

```python
df = pd.read_csv('private_data/cibc_visa_2023_05_03_2022_01_01.csv', header=None, names=['date', 'description', 'debit', 'credit', 'expender'])
```

```python
df.loc[df['expender'].str.endswith('6875'), 'expender'] = 'yousuf'
df.loc[df['expender'].str.endswith('5826'), 'expender'] = 'nisreen'
```

```python
df = df.fillna(0)
```

```python
df['expense'] = df['debit'] - df['credit']
```

```python
df['type'] = 'debit'
df.loc[df['credit'] > 0, 'type'] = 'credit'
```

```python
df['month'] = pd.to_datetime(df['date']).dt.to_period('M')
```

```python
df.groupby(['month']).sum()
```

```python
df.groupby(['description']).sum().sort_values('expense', ascending=False).head(30)
```

```python
df.groupby(['description', 'month']).sum()
```

```python
df.pivot_table(index='description', columns='month', values='debit', fill_value=0)
```
